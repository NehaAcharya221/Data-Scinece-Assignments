#### Q2.Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in  Jupyter Notebook. 

```python
# code to display "hello world!!"
from flask import Flask

app = Flask(__name__)

@app.route('/')
def display_hello():
    return 'Hello World!!'

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

#### Q3.What is App routing in Flask? Why do we use app routes? 

#### Answer: Flask App Routing means mapping the URLs to a specific function that will handle the logic for that URL. Modern web frameworks use more meaningful URLs to help users remember the URLs and make navigation simpler. Example: In our application, the URL (“/”) is associated with the root URL.So if our site’s domain was www.pwskills.org and we want to add routing to “www.pwskills.org/hello”, we would use “/hello”. 

#### App routing is used to map the specific URL with the associated function that is intended to perform some task. The routing technique helps a user remember application URLs and is useful to access the desired page directly without having to navigate from the home page.

#### Q4.Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”  route to show the following details: 
####    Company Name: ABC Corporation 
 ####       Location: India 
 ####       Contact Detail: 999-999-9999  
####    Attach the screenshot of the output in Jupyter Notebook.

```python
# code o display welcome and details.
from flask import Flask

app = Flask(__name__)

@app.route('/')
def display_details():
    return '<p>Company Name: ABC Corporation</p><p>Location: India</p><p>Contact Detail: 999-999-9999</p>'

@app.route("/welcome")
def welcome():
    return "Welcome to ABC Corporation"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

#### Q5.What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the  url_for() function. 

#### Aswner: url_for() function is used in flask for url building.It takes name of the function as argument & returns the URL associated with that function.

```python
# code for url building
from flask import Flask
from flask import url_for

app = Flask(__name__)

@app.route('/')
def display_hey():
    return "Hey World!!!!"

with app.test_request_context():
    print(url_for('display_hey'))

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
