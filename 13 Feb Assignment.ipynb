####  Q4.Why LookupError class is used? Explain with an example KeyError and IndexError

#### Answer: Lookup Error acts as a base class for the exceptions that occur when a key or index used on a mapping or sequence of a list/dictionary is invalid or does not exists. The two types of exceptions raised are: IndexError. KeyError.

```python
# Example of IndexError 
# index 10 is not available in the list l so it will print IndexError as List index out of range
try:
    l=[1,2,3,5,6]
    l[10]
except IndexError as f : 
    print(f)
```

```python
# Example of KeyError 
# key10 is not available in the dict d so it will print KeyError as key10 is not available in dict
try:
    d = {1:[3,4,5,6],"key":"sudh"}
    d["key10"]
except KeyError as e:  
    print("key10 is not available in dict")
```

#### Q5.Explain ImportError. What is ModuleNotFoundError? 

#### ANswer: ImportError generally occurs when a class cannot be imported due to one of the following reasons: The imported class is in a circular dependency. The imported class is unavailable or was not created. The imported class name is misspelled. The imported class from a module is misplaced.
### Below is the example of ImportError

```python
# example of ImportError
# as neha module does not exists it will print as No module named 'neha'
try :
    import neha
except ImportError as e: 
    print(e)
```

#### ModuleNotFoundError =This error occurs when you're trying to access or use a module that cannot be found.
